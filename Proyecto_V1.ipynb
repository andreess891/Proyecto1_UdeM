##  Descripci贸n de los Datos

Para el entrenamiento del modelo supervisado de clasificaci贸n se cuenta con un conjunto de datos hist贸ricos que contiene informaci贸n sobre 贸rdenes de desviaci贸n significativa generadas por incrementos en el consumo de servicios p煤blicos. Estas 贸rdenes fueron previamente revisadas por analistas, quienes tomaron decisiones basadas en las condiciones observadas.

###  Variable Objetivo: `RESPUESTA`

La variable `RESPUESTA` representa la decisi贸n final tomada por el analista, y se comporta de la siguiente manera:

| Caso | 驴Modifica Consumo? | 驴Env铆a a terreno? | RESPUESTA |
|------|--------------------|-------------------|-----------|
| 1    | S铆                 | No                | 1         |
| 2    | S铆                 | S铆                | 1         |
| 3    | No                 | S铆                | 1         |
| 4    | No                 | No                | 0         |

####  Interpretaci贸n de los casos:
- **Caso 1:** El analista detecta un error claro en el consumo y lo ajusta sin necesidad de verificaci贸n en terreno.
- **Caso 2:** El analista ajusta parcialmente el consumo, pero requiere una visita en terreno para validar la parte restante antes de decidir su facturaci贸n definitiva.
- **Caso 3:** El analista no ajusta el consumo por falta de evidencia suficiente, pero mantiene dudas que requieren una inspecci贸n en terreno.
- **Caso 4:** El consumo investigado es considerado v谩lido y no necesita ajustes ni verificaci贸n en terreno.

Por tanto, la clase `1` agrupa todas las situaciones en las que se toma alguna acci贸n correctiva o preventiva adicional, mientras que la clase `0` representa los casos cerrados sin intervenci贸n alguna.

---

### Ь Variables Predictoras

Las dem谩s columnas del dataset representan condiciones observadas por el analista en el momento de tomar la decisi贸n, y ser谩n utilizadas como **variables predictoras** en el modelo. Estas incluyen, entre otras:

- `CONSUMO_CRITICADO`: Valor del consumo que se encuentra en investigaci贸n.
- `SERVICIO`: Tipo de servicio involucrado (agua, energ铆a, etc.).
- `CATEGORIA`: Segmento del usuario (residencial, industrial, etc.).
- `NIVEL_TENSION`: Capacidad de energ铆a entrega y solo aplica para el servicio de energ铆a
- `ESTRATO`: Estrato socio economico del predio que de 1 a 6 y solo aplica para el sector residencial (CATEGORIA: Residencial)
- `LOCALIDAD`: Municipio donde se se presta el servicio.
- `FUNCION_ANALISIS`: Algoritmo o l贸gica aplicada para el an谩lisis del consumo.
- `CALIFICACION`: Clasificaci贸n del nivel de desviaci贸n identificado.
- `OBS_LECTURA`: Observaciones sobre la lectura del medidor que realiza el lector.
- `PERIODICIDAD`: Frecuencia con la que se factura el servicio.

**Cantidad de observaciones**: 85.812

Estas variables permitir谩n al modelo aprender de los patrones y comportamientos hist贸ricos, con el objetivo de replicar las decisiones de los analistas en nuevos casos.

```python
# Analizar la variable 'ESTRATO'
estrato_counts = df['estrato'].value_counts()

# Analizar la variable  'SERVICIO' 
servicio_counts = df['servicio'].value_counts()

# Relacion entre ESTRATO y SERVICIO
contingency_table = pd.crosstab(df['estrato'], df['servicio'])

# Chi.cuadrado
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Graficar los resultados
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 12))

# Por ESTRATO
axes[0].bar(estrato_counts.index.astype(str), estrato_counts.values)
axes[0].set_title('Distribuci贸n de ESTRATO')
axes[0].set_xlabel('ESTRATO')
axes[0].set_ylabel('Frecuencia')

# Por SERVICIO
axes[1].bar(servicio_counts.index.astype(str), servicio_counts.values)
axes[1].set_title('Distribuci贸n de SERVICIO')
axes[1].set_xlabel('SERVICIO')
axes[1].set_ylabel('Frecuencia')
axes[1].tick_params(axis='x', rotation=25)

# Relacion de estrato y servicio
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='YlGnBu', ax=axes[2])
axes[2].set_title('Tabla de relaci贸n entre ESTRATO y SERVICIO')
axes[2].set_xlabel('SERVICIO')
axes[2].set_ylabel('ESTRATO')

plt.tight_layout()
plt.show()

print("An谩lisis de la variable ESTRATO:")
print(estrato_counts)
print("\nAn谩lisis de la variable SERVICIO:")
print(servicio_counts)
print("\nTabla de relaci贸n entre ESTRATO y SERVICIO:")
print(contingency_table)
print("\nResultados del test Chi-cuadrado:")
print(f"Chi-cuadrado: {chi2}")
print(f"P-valor: {p}")
print(f"Grados de libertad: {dof}")

```

Conclusiones:
- El estrato 3 destaca por su mayor presencia en el conjunto de datos, seguido por el estrato 2, lo que sugiere una desviaci贸n m谩s pronunciada en el uso de los servicios p煤blicos dentro de estos grupos socioecon贸micos.
- El estrato 6 presenta una menor participaci贸n, lo que sugiere que las desviaciones en el consumo no ocurren con frecuencia en este grupo socioecon贸mico.
- El servicio que m谩s presenta consumos at铆picos es el agua potable, esto se puede presentar por fugas imperceptibles, seguido de la energ铆a.
